## TextBlob - PatternAnalyzer

`TextBlob(text,analyzer=PatternAnalyzer()).sentiment` returns you the polarity score which is a float within the range [-1.0, 1.0]. -1 means negative sentence or 1.0 positive sentence. The subjectivity is a float value within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.

* polarity > 0.2 - positive
* polarity < 0.2 - negative

```python
for i, text in enumerate(data):
    print(f" * sentece: {i + 1}: {text}")
    print(">>>", TextBlob(text, analyzer=PatternAnalyzer()).sentiment, "\n")
```

## TextBlob - NaiveBayesAnalyzer

`NaiveBayesAnalyzer` which returns you the sentiment of the given text whether it is positive or negative to get the polarity of the sentence 

```python
for i, text in enumerate(data):
    print(f" * sentece: {i + 1}: {text}")
    print(">>>", TextBlob(text, analyzer=NaiveBayesAnalyzer()).sentiment, "\n")
```

# Vader

`Vader`: (Valence Aware Dictionary and sEntiment Reasoner) It is a python library focused on social media sentiment, is an open-sourced package within the `Natural Language Toolkit (NLTK)` and is the best pre-trained sentiment classifier for social media conversations from networks such as Facebook or Twitter. It uses a combination of sentiment lexicon is a list of lexical features (e.g., words) which are generally labeled according to their semantic orientation as either positive or negative. VADER not only tells us about the Positivity and Negativity score but also tells us about how positive or negative sentiment is. It’s less accurate when rating longer, structured sentences, but it’s often a good launching point.

```python
!pip install vaderSentiment ## install the library if you don't have
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentiment = SentimentIntensityAnalyzer()
```

The `compound` score is computed by summing the valence scores of each word in the lexicon, adjusted according to the rules, and then normalized to be between -1 (most extreme negative) and +1 (most extreme positive). This is the most useful metric if you want a single unidimensional measure of sentiment for a given sentence. Calling it a ‘normalized, weighted composite score’ is accurate.

* positive sentiment: compound score >= 0.05
* neutral sentiment: (compound score > -0.05) and (compound score < 0.05)
* negative sentiment: compound score <= -0.05

```python
for i, text in enumerate(data):
    print(f" * sentece: {i + 1}: {text}")
    print(">>>", sentiment.polarity_scores(text), "\n")
```

## bert-base-uncased

```python
def sentiment_labels_update(text):
    # Move model to eval mode for consistency
    model.eval() 
    
    encoded_input = tokenizer(text, padding=True, truncation=True, max_length=512, return_tensors='pt')
    
    with torch.no_grad(): # Saves memory/compute by not tracking gradients
        output = model(**encoded_input)
    
    # Use .squeeze() to remove the batch dimension safely
    scores = output.logits.squeeze().numpy() 
    scores = softmax(scores)
    
    predicted_idx = np.argmax(scores)
    return config.id2label[predicted_idx]
```

```python
MODEL = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(MODEL)
config = BertConfig.from_pretrained(MODEL)
```

```python
# Manually define the mapping
config.id2label = {0: "NEGATIVE", 1: "POSITIVE"}
config.label2id = {"NEGATIVE": 0, "POSITIVE": 1}
model = BertForSequenceClassification.from_pretrained(MODEL, config=config)
```

```python
for i, text in enumerate(data):
    print(f" * sentece: {i + 1}: {text}")
    print(">>>", sentiment_labels_update(text), "\n")
```

# Zero Shot Classification

`Zero Shot Classification`: Zero-Shot Classification is the task of predicting a class that wasn’t seen by the model during training. This method, which leverages a pre-trained language model, can be thought of as an instance of [`transfer learning`](https://www.youtube.com/watch?v=BqqfQnyjmgg) which generally refers to using a model trained for one task in a different application than what it was originally trained for. This is particularly useful for situations where the amount of labeled data is small. Refer this [hugging face website](https://huggingface.co/tasks/zero-shot-classification) to know more about this.

```python
from transformers import pipeline
classifier = pipeline(
                      task="zero-shot-classification",
#                       device=0,
                      model="facebook/bart-large-mnli"
                    )
```

```python
for i, text in enumerate(data):
    print(classifier(text,["positive","negative",'neutral'],multi_class=True), "\n")
```
